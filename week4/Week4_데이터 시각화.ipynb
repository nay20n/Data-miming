### **1.3 Matplotlib 사용 방법**

#### **1.3.1 핵심 모듈:** `pyplot`

* `matplotlib.pyplot`은 MATLAB 스타일의 인터페이스 제공

* 자동으로 그림(Figure)과 축(Axes)을 생성해 초보자에게 적합

* 복잡한 시각화(예: 여러 subplot)에는 객체지향 방식(Figure, Axes 직접 생성) 권장

#### **1.3.2 시각화 기본 절차**

1. 데이터 준비

2. pyplot 모듈 임포트

3. 시각화 함수 호출 (예: `plt.plot()`)

4. 제목/라벨/범례 설정

5. `plt.show()` 호출

```python
# 데이터 준비
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# pyplot 모듈 임포트
import matplotlib.pyplot as plt

# 시각화 함수 호출
plt.plot(x, y)

# 제목/라벨/범례 설정
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# plt.show() 호출
plt.show()
```

### **1.4 자주 사용하는 그래프 종류**

| 그래프 종류         | 함수                      | 사용 예                         |
|--------------------|---------------------------|----------------------------------|
| 선 그래프 (Line)   | `plt.plot()`              | 시계열 변화 시각화              |
| 막대 그래프 (Bar)  | `plt.bar()`, `plt.barh()` | 범주형 데이터 비교               |
| 히스토그램         | `plt.hist()`              | 데이터 분포 확인                 |
| 산점도             | `plt.scatter()`           | 두 변수 간의 상관관계 시각화     |
| 파이 차트          | `plt.pie()`               | 전체 중 각 항목의 비율 표현      |
| 박스플롯           | `plt.boxplot()`           | 이상치 탐색 및 데이터 분포 확인  |
| 히트맵 (Heatmap)  | `plt.imshow()`           | 행렬형 데이터 시각화 (예: 상관계수)  |

## **2. 단일 변수의 시각화**

### **2.1 히스토그램(Histogram)**

* 히스토그램은 데이터 분포를 직관적으로 이해할 수 있는 중요한 시각화 도구임

* 막대 그래프 형태로 데이터를 급간(bin)으로 나누고, 각 급간에 속하는 데이터의 빈도수를 막대의 높이로 표현함

* **급간(bin) 선택의 중요성:**
    * 급간의 개수는 히스토그램 해석에 큰 영향을 미치므로 적절한 설정이 중요함
    * 데이터 크기에 비해 너무 많거나 적은 급간을 사용하면 의미 있는 분석 결과를 도출하는 데 한계가 있음
    * 일반적으로 데이터의 양과 복잡성에 따라 5개에서 15개 급간을 사용하는 것을 권장하며, 데이터가 많고 다양할수록 더 많은 급간을 고려할 수 있음

#### **예제: 국가별 교통사고 사망률**

* 아래의 데이터는 30개 국가의 1980년 중반 교통사고 사망률(인구 10만 명당)을 나타냄

```python
# Pandas 라이브러리 호출
import pandas as pd

# 예제 데이터 불러오기
traffic = pd.read_csv("deathrates.csv")

# 데이터프레임의 크기 출력
print("데이터프레임의 크기:", traffic.shape)

# 데이터프레임의 처음 5개 행 출력
traffic.head()
```

* 히스토그램은 `plt.hist()` 함수를 사용하여 빠르고 효율적으로 만들 수 있음

```python
# 히스토그램 생성
plt.hist(traffic["death"])

# 그래프 표시
plt.show()
```

* `plt.hist()` 함수는 구간 개수, 막대 색상 등 다양한 옵션을 분석 목적에 맞게 설정할 수 있음
  
    * `bins`: 급간 개수
    
    * `color`: 막대 색상
    
    * `alpha`: 투명도
    
    * `rwidth`: 막대 너비 비율
    
    * `edgecolor`: 테두리 색상
   
    * `linewidth`: 테두리 선 두께

```python
# 급간, 색상 등 옵션 지정
plt.hist(traffic["death"], bins=6, color= "lightblue", alpha=0.7, rwidth=0.9, edgecolor="black", linewidth=0.2)

# 그래프 표시
plt.show()
```

* 모든 그래프는 색상이나 제목 등을 꾸미기 위한 다양한 인자를 사용할 수 있음
  
  * `plt.figure()`: 그래프 전체 틀 설정 (크기, 해상도, 배경색 등)

  * `plt.title()`: 그래프 제목 설정 (텍스트, 크기, 위치, 색상 등)
 
  * `plt.xlabel()` / `plt.ylabel()`: 축 이름 지정
 
  * `plt.ylim()` / `plt.xlim()`: 축 범위 설정

```python
# 그래프 크기 설정
plt.figure(figsize=(10, 5), dpi=100, facecolor='white', edgecolor='black')

# 급간, 색상 등 인자 지정
plt.hist(traffic["death"], bins=6, color= "lightblue")

# 제목 설정
plt.title('Traffic Death Rates', fontsize=14, fontweight='bold', loc='center', pad=20)

# x축, y축 레이블 설정
plt.xlabel('Death Rates', fontsize=12, fontweight='bold', labelpad=10, color='black')         
plt.ylabel('Frequency', fontsize=12, fontweight='bold', labelpad=10, color='black')   

# y축 범위 설정
plt.ylim(0, 20)                                                   

# 그래프 표시
plt.show()
```

#### **색상의 사용**
* Matplotlib에서 색상은 이름을 이용해 쉽게 지정할 수 있으며, 그래프의 다양한 요소(선, 텍스트, 축 등)에 적용 가능함

* `matplotlib.colors` 모듈의 CSS4_COLORS 사전에서 사용 가능한 모든 색상 이름과 해당 RGB 값들을 확인할 수 있음 → 공식 색상 목록 참고: https://matplotlib.org/stable/gallery/color/named_colors.html


```python
from matplotlib import colors

# 사용 가능한 모든 색상 이름 출력
for name, hex in colors.CSS4_COLORS.items():
    print(name, hex)
```

### **2.2 막대그래프**

* 막대그래프는 빈도 분포표를 시각적으로 표현하는 데 사용되는 대표적인 그래프로, 이산형(discrete) 변수나 범주형(categorical) 변수의 데이터 분포를 나타낼 때 유용함

* 각 범주의 데이터 값이나 빈도 수를 막대의 높이로 표시하여 데이터 간의 비교를 직관적으로 할 수 있음

#### **예제: 도시 인구 변화**

* "popcities10.csv" 데이터 파일은 한국의 여러 도시들에 대한 2000년, 2005년, 2010년의 인구를 나타냄

```python
# 예제 데이터 불러오기
pop = pd.read_csv("popcities10.csv")

# 데이터프레임의 크기 출력
print(pop.shape)

# 데이터프레임의 처음 5개 행 출력
pop.head()
```

* 막대그래프는 `plt.bar()` 함수를 사용하여 쉽게 만들 수 있음

  * 예제에서는 인구 수를 기준으로 상위 10개 도시를 선택하고, 각 도시의 2010년 인구 데이터를 사용함

```python
# 그래프 크기 설정
plt.figure(figsize=(12, 8))

# 막대 그래프 생성
plt.bar(pop['municipality'][:10], pop['pop10'][:10])

# 그래프 표시
plt.show()
```

* y축의 값이 지나치게 클 경우에는 단위를 변경하여 가독성을 높이는 것이 효과적임 (예: 1,000명 단위로 환산)

```python
# 그래프 크기 설정
plt.figure(figsize=(12, 8))

# y축 인구 수 단위를 변경
plt.bar(pop['municipality'][:10], pop['pop10'][:10] / 1000)

# 제목 및 축 라벨 추가 
plt.title('Population Change in Selected Cities')
plt.xlabel('City')
plt.ylabel('Population (in thousands)')

# 그래프 표시
plt.show()
```

* 막대그래프는 세로 방향뿐 아니라 `plt.barh()` 함수를 통해 수평 막대그래프로도 표현 가능함

```python
# 그래프 크기 설정
plt.figure(figsize=(12, 8))

# 막대 그래프 생성
plt.barh(pop['municipality'][:10], pop['pop10'][:10] / 1000)

# 제목 및 축 라벨 추가 
plt.title('Population Change in Selected Cities')
plt.xlabel('Population (in thousands)')
plt.ylabel('City')

# 그래프 표시
plt.show()
```

* 하나의 범주에 대해 여러 시점 데이터를 비교해야 할 경우에는는 그룹 형태의 막대그래프로 시각화할 수 있음 → 각 그룹마다 시점별로 막대를 나란히 배치하여 연도별 추세를 비교함

```python
import numpy as np

# 그래프 크기 설정
plt.figure(figsize=(12, 8))

# 도시 선택
selected_cities = pop['municipality'][:10]
pop_2000 = pop['pop00'][:10] / 1000
pop_2005 = pop['pop05'][:10] / 1000
pop_2010 = pop['pop10'][:10] / 1000

# X축의 위치 설정
x = np.arange(len(selected_cities))

# 막대그래프 그리기
plt.bar(x, pop_2000, width=0.2, label='2000 years')
plt.bar(x + 0.2, pop_2005, width=0.2, label='2005 years')
plt.bar(x + 0.4, pop_2010, width=0.2, label='2010 years')

# X축에 도시 이름 표시
plt.xticks(x + 0.2, selected_cities)

# 범례 표시
plt.legend()

# 제목 및 축 라벨 추가
plt.title('Population Change in Selected Cities')
plt.xlabel('Population (in thousands)')
plt.ylabel('City')

# 그래프 표시
plt.show()
```

## **3. 여러 변수의 시각화**

### **3.1 선 그래프**

* 선 그래프는 변화 추세나 경향을 나타내는 데 유용한 시각화 도구로, 일반적으로 시간에 따른 연속적인 값을 시각화할 때 사용됨

* 연도별 변화, 누적 값, 시간 흐름에 따른 데이터 패턴 등을 파악하는 데 효과적임

#### **예제: 도시 인구 변화**

* 아래 예제 데이터는 한국의 여러 도시들에 대한 연도별 인구 변화를 나타내고 있음

```python
# 예제 데이터 불러오기
pop = pd.read_csv("popcities10.csv")

# 데이터프레임의 크기 출력
print(pop.shape)

# 데이터프레임의 처음 5개 행 출력
pop.head()
```

* `plt.plot()` 함수를 이용하여 선 그래프를 쉽게 생성할 수 있으며, 선의 모양, 마커(점), 색상 등을 자유롭게 조정할 수 있음

* 마커 스타일과 선 유형은 아래 공식 문서를 통해 확인할 수 있음
  
  * `marker`: https://matplotlib.org/stable/api/markers_api.html 
  
  * `linestyle`: https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle 

```python
# plt.plot() 함수수
plt.plot([2000, 2005, 2010], pop.iloc[0, 3:6].sort_index() / 1000, marker='o', linestyle='--', color='b')
```

* 여러 도시의 선 그래프를 동시에 출력하면 도시간 인구 증가 추이를 직관적으로 비교할 수 있음
  * `label` 인자를 사용하여 도시명을 지정하고 `plt.legend()`를 활용하면 범례를 통해 그래프 해석을 쉽게 할 수 있음

```python
# 여러 선 그래프 생성
plt.plot([2000, 2005, 2010], pop.iloc[2, 3:6].sort_index() / 1000, marker='o', linestyle='--', color='b', label='Incheon')
plt.plot([2000, 2005, 2010], pop.iloc[3, 3:6].sort_index() / 1000, marker='o', linestyle='--', color='b', label='Daegu')

# 범례 추가
plt.legend(facecolor='white', edgecolor='black', loc='upper left')
```

*  `plt.plot()` 함수를 사용할 때 `marker`와 `linestyle` 옵션을을 활용하면 데이터 포인트의 표현 방식과 이들을 연결하는 선의 스타일을 다양하게 조정할 수 있음
    
    * `marker`: 데이터 포인트 각각에 적용될 마커(점 기호)의 유형을 지정
    
    * `linestyle`: 데이터 포인트를 연결하는 선의 스타일을 지정
    
    * 공식 문서 참고: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# 샘플 데이터
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 산점도 및 선 그래프 생성
plt.plot(x, y, marker='o', linestyle='solid', label='Solid Line')
plt.plot(x, [i+1 for i in y], marker='s', linestyle='dotted', label='Dotted Line')
plt.plot(x, [i+2 for i in y], marker='^', linestyle='dashed', label='Dashed Line')
plt.plot(x, [i+3 for i in y], marker='*', linestyle='dashdot', label='Dash-dot Line')

# 범례 표시
plt.legend()

# 그래프 제목 및 축 라벨 설정
plt.title('Different Marker and Linestyle')
plt.xlabel('X axis')
plt.ylabel('Y axis')

# 그래프 표시
plt.show()
```

### **3.2 산점도**

* 산점도는 두 변수 간의 관계를 시각적으로 파악할 수 있는 대표적인 방법임

* 각 데이터 포인트를 x축과 y축 좌표에 맞게 점으로 표시하며, 변수 간의 관계 패턴, 이상치 여부 등을 탐색하는 데 적합함

* 산점도를 통해 두 변수 간의 방향성, 선형성, 관계 강도를 확인할 수 있으며, 전반적인 패턴에서 벗어난 이상치(outlier) 도 쉽게 식별할 수 있음

#### **예제: 항공사 아웃소싱** 

* 아래 데이터는 각 항공사의 아웃소싱 비율과 그에 따른 운항 지연 비율을 나타냄

```python
# 예제 데이터 불러오기
flights = pd.read_csv("outsourcing.csv")

# 데이터프레임의 크기 출력
print(flights.shape)

# 데이터프레임의 처음 5개 행 출력
flights.head()
```

* `plt.scatter()` 함수를 사용하여 산점도를 생성할 수 있으며, 다양한 마커나 색상, 크기 조절이 가능함

```python
# scatter() 함수 사용
plt.scatter(flights["outsourcing"], flights["delays"], marker='v', s=50, c='r', alpha=0.5)

# 그래프 표시
plt.show()
```

* `plt.axhline()` 함수는 그래프에 수평선을 추가할 수 있는 도구로, 특정 기준선이나 임계값을 시각적으로 강조할 때 유용함

* 선의 색상, 스타일, 두께 등 다양한 속성을 설정할 수 있음 → 공식 문서 참고: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axhline.html 

```python
# 산점도 생성
plt.scatter(flights["outsourcing"], flights["delays"], marker='v')

# 제목 및 축 라벨 추가
plt.title('Outsourcing vs. Flight Delays')
plt.xlabel('Percentage of Outsourced Maintenance')
plt.ylabel('Percentage of Flight Delays')

# 수평선 추가
plt.axhline(y=15, color='b', linestyle='-', linewidth=1)
plt.axhline(y=20, color='g', linestyle='--', linewidth=2)
plt.axhline(y=25, color='r', linestyle='-.', linewidth=3)
plt.axhline(y=30, color='m', linestyle=':', linewidth=4)

plt.show()
```

* `plt.plot()` 함수도 `linestyle='None'`과 `marker` 인자를 함께 사용하면 산점도 형태의 시각화가 가능함

```python
# plot() 함수 사용
plt.plot(flights["outsourcing"], flights["delays"], marker = 'o', linestyle='None', markersize=5)  # 'o' 마커를 사용하여 점으로 표시

# 그래프 표시
plt.show()
```

* 두 함수 모두 산점도를 시각화할 수 있지만, 목적과 스타일 설정 방식에서 차이가 있음

| 항목 | `plt.plot()` | `plt.scatter()` |
|------|--------------|------------------|
| **주 목적** | 선 그래프 / 선으로 연결된 점 표현 | 산점도 (독립적인 점 시각화) |
| **기본 표현** | 점을 선으로 연결 (기본: `linestyle='-'`) | 점만 표현 (선 없음) |
| **선 연결 여부** | 선 있음 (`linestyle` 설정 가능) | 선 없음 (각 점은 독립적) |
| **사용 상황** | 시계열, 함수 표현 등 연속형 데이터 | 두 변수 간 관계 또는 분포 분석 |
| **스타일 조정** | 선 스타일, 마커, 색상 일괄 설정 | 점 크기(`s`), 색상(`c`) 등 개별 설정 가능 |
| **대표 옵션** | `linestyle`, `marker`, `color` | `s`, `c`, `alpha`, `edgecolors` 등 |
