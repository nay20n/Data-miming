### **실습 문제**

국토교통부에서 운영하는 실거래가 공개시스템(http://rt.molit.go.kr/)은 2006년 1월 이후 체결된 부동산 거래 정보를 제공하고 있습니다. 이번 실습에서는 실거래가 공개시스템에서 다운로드한 데이터를 이용하여 데이터 탐색 및 전처리 함수를 학습해보겠습니다.
e-class의 데이터 압축 파일에는 2018년 상반기 동대문구와 성북구의 아파트 및 단독주택 매매 거래 정보가 포함되어 있으며, 건축 연도나 면적 등의 기본 정보는 별도 파일로 저장되어 있습니다(총 4개의 CSV 파일).

각 파일은 다음과 같이 Pandas 라이브러리의 read_csv() 함수를 이용하여 불러오겠습니다.

```python
import pandas as pd

aptsales = pd.read_csv("apt_sales.csv", encoding="CP949")          # 아파트 매매거래 정보
aptinfo = pd.read_csv("apt_basicinfo.csv", encoding="CP949")       # 아파트 기본 정보
housesales = pd.read_csv("house_sales.csv", encoding="CP949")      # 단독주택 매매거래 정보
houseinfo = pd.read_csv("house_basicinfo.csv", encoding="CP949")   # 단독주택 기본 정보
```

##### 1. [10점] 단독주택 매매거래 정보를 사용하여 다음의 질문에 답하세요. 모든 답안은 실행 가능한 코드로 작성하고 코드로 작성할 수 없는 설명은 반드시 주석(##)으로 작성하여 실행 시 오류가 발생하지 않도록 하세요.

(1) [2점] info( ) 함수를 사용하여 단독주택 매매거래 데이터에 총 몇 개의 변수가 저장되어 있는지 확인하고, 그 중 매매가격(salesPrice)은 몇 번째에 위치해 있는지 적으시오.

```python
housesales.info() 
# Data columns (total 8 columns) 정보에 따르면 총 변수 개수는 8개이고 매매가격위치는 7번째에 위치해 있다
```

(2) [2점] merge() 함수를 사용하여 단독주택 매매거래 데이터와 기본 정보 데이터를 id 변수를 기준으로 결합하고 이를 새로운 데이터프레임으로 저장하는 코드를 작성하시오.

```python
df = pd.merge(housesales, houseinfo, on='id', how='inner')
df
```

(3) [2점] 결합 후 중복되는 변수는 제거한 후 rename() 함수를 사용하여 단독주택 데이터의 변수명을 si는 city, gu는 district, dong은 street, bunji는 number로 변경하는 코드를 작성하시오.

```python
df=df.rename(columns={'si_x': 'city_x', 'si_y': 'city_y', 'gu_x': 'district_x', 'gu_y': 'district_y', 'dong_x': 'street_x', 'dong_y': 'street_y', 'bunji_x': 'number_x', 'bunji_y': 'number_y',})
df
```

(4) [2점] 단독주택 매매거래 데이터에서 단위면적(landSize)당 매매가격을 나타내는 새로운 파생변수가 추가된 데이터프레임을 생성하는 코드를 작성하시오.

```python
df['price_per_unit'] = df['salesPrice'] / df['landSize']
```

(5) [2점] 성북구에 소재한 단독주택의 단위면적당 매매가격 분포를 describe() 함수를 이용하여 확인하는 코드를 작성하시오.

```python
df.describe()
```

##### 2. [15점] 아파트 매매거래 데이터를 사용하여 다음 질문에 답하세요.모든 답안은 실행 가능한 코드로 작성하고 코드로 작성할 수 없는 설명은 반드시 주석(##)으로 처리하여 실행 시 오류가 발생하지 않도록 하세요

(1) [2점] info( ) 함수를 사용하여 아파트 매매거래 데이터에 총 몇 개의 변수가 저장되어 있는지 확인하고, 그 중 매매가격(salesPrice)은 몇 번째에 위치해 있는지 적으시오.

```python
aptsales.info() #매매가격은 9번째에 위치하고 변수는 총 10개가 저장되어있다.
```

(2) [2점] merge() 함수를 사용하여 아파트 매매거래 데이터와 기본 정보 데이터를 id 변수를 기준으로 결합하고 이를 새로운 데이터프레임으로 저장하는 코드를 작성하시오.

```python
df2 = pd.merge(aptsales,aptinfo, on='id', how='inner') #inner교집합 , 공통된 id값만 변합된다.
```

(3) [2점] 결합 후 중복되는 변수는 제거한 후 rename() 함수를 사용하여 아파트 데이터의 변수명을 si는 city, gu는 district, dong은 street, bunji는 number로 변경하는 코드를 작성하시오.

```python
df2=df2.rename(columns={'si_x': 'city_x', 'si_y': 'city_y', 'gu_x': 'district_x', 'gu_y': 'district_y', 'dong_x': 'street_x', 'dong_y': 'street_y', 'bunji_x': 'number_x', 'bunji_y': 'number_y',})
df2
```

(4) [2점] 아파트 데이터에서 단위면적(size) 당 매매가격을 나타내는 새로운 파생변수가 추가된 데이터프레임을 생성하는 코드를 작성하시오.

```python
 df2['price_per_unit'] = df2['salesPrice'] / df2['size'] #가격/단위면적을 하게 되면 된다.
```

(5) [2점] describe() 함수를 사용하여 동대문구에 소재한 아파트의 단위면적당 매매가격 분포를 확인하는 코드를 작성하시오.

```python
df2.describe() 
```

(6) [3점] cut() 함수를 사용하여 salesPrice 변수를 4개 구간으로 나누고, 각 구간에 따라 "A", "B", "C", "D" 값을 부여하는 코드를 작성하시오.

```python
# bin 자동으로 4개의 구간으로 나누어 오름차순으로 D C B A로 값을 부여한다
df2['price_category'] = pd.cut(df2['salesPrice'], bins=4, labels=["D", "C", "B", "A"])
df2
```

(7) [2점] 건축연도(yearBuilt)를 기준으로 2000년 이전은 "0", 2000년 이후는 "1"로 구분하는 더미변수를 생성하고 저장하는 코드를 작성하시오.

```python
# 0~1999년까지는 0, 2000~3000년까지는 1로 구분한다.
bins = [0, 1999, 3000]
df2['yearBuilt_dummy'] = pd.cut(df2['yearBuilt'], bins, labels=[0, 1])
df2
```
