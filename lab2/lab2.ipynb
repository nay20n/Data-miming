##### 1. [10점] ‘deathrates.csv’ 파일에는 1950년대 중반 30개 국가의 인구 10만 명당 교통사고 사망률 데이터가 포함되어 있습니다. 아래 질문에 대한 답변 코드를 작성하시오.

(1) [2점] `hist()` 함수를 사용하여 교통사고 사망률에 대한 히스토그램을 출력하는 코드를 작성하시오. 그래프에는 제목, x축 및 y축에 대한 레이블이 포함되어야 하며 나머지 부분은 기본값을 사용하면 됩니다.

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')

plt.figure(figsize=(10,5))
plt.hist(deathrates['death'],bins=6,color='blue')
plt.title("Death Rates",pad=20)
plt.xlabel("Death Rate")
plt.ylabel("Frequency")
plt.ylim(0,20) #y축 그래프 0~20

plt.show()

#hist함수 
#deathrates.head(10)
#deathrates.info()
#공백있음 히스토그램 x
```

(2) [2점] `bar()` 함수를 사용하여 국가별 교통사고 사망률을 오름차순으로 정렬하여 막대그래프로 나타내는 코드를 작성하시오. 그래프에는 제목, x축 및 y축의 값을 설명하는 레이블이 포함되어야 하며 나머지 부분은 기본값을 사용하면 됩니다.

```python
deathrates_sort = deathrates.sort_values(by="death")
deathrates_sort
```

```python
plt.figure(figsize=(12,8))
plt.bar(deathrates_sort["country"],deathrates_sort["death"])
plt.xticks(rotation=75) 

plt.show()
```

(3) [2점] `cut()` 함수를 사용하여 전체 교통사고 사망률을 ‘Low’, ‘Medium’, ‘High’의 세 가지로 분류하는 코드를 작성하시오. 분류하는 기준은 임의로 설정할 수 있습니다.

```python
# 답안을 작성하세요.
deathrates["death"].describe()
```

```python
death_label = pd.cut(deathrates["death"], bins=[0,12,13,27], labels = ["Low", "Medium", "High"])
```

(4) [4점] (3)에서 가공된 교통사고 데이터를 사용하여 결과를 그림 1과 같이 막대그래프로 표현하는 코드를 작성하시오. 그래프에는 제목, x축 및 y축의 값을 설명하는 레이블이 포함되어야 하며 나머지 부분은 기본값을 사용하면 됩니다.

```python
# 답안을 작성하세요.
death_label_sort=death_label.value_counts().reindex(["Low","Medium","High"])
death_label_sort
```

```python
death_label_sort.index
```

```python
death_label_sort.values
```

```python
plt.figure(figsize=[12,8])
plt.bar(death_label_sort.index,death_label_sort.values)
plt.style.use('ggplot')
#plt.grid(True)
plt.title("Traffic fatalities in each country", fontsize=18)
plt.xlabel("Tracffic death rates")
plt.ylabel("Number of countries", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

##### 2. [15점] 'popcities10.csv' 파일에는 한국 내 83개 도시의 2000년, 2005년, 2010년 인구 데이터가 포함되어 있습니다. 아래 질문에 대한 답변 코드를 작성하시오.

(1) [3점] `plot()` 함수를 사용하여 2000부터 2010년까지 전라북도 군산의 인구 변화와 해당 기간 평균 인구를 그림 2와 같이 보여주는 그래프를 나타내는 코드를 작성하시오.

```python
gunsan_data = pop[pop["municipality"] == "Gunsan"]

pop_xlabel = [2000,2005,2010]
pop_ylabel = gunsan_data.iloc[0][["pop00", "pop05", "pop10"]].astype(int).tolist()

pop_mean = sum(pop_ylabel) / len(pop_ylabel)
```

```python
plt.figure(figsize=[12,8])
plt.plot(pop_xlabel,pop_ylabel, marker='o', linewidth=2.5)
plt.axhline(y=pop_mean, linestyle='--', linewidth=2.5) 
plt.title("Population of Gunsan")
plt.xlabel("Year")
plt.ylabel("Population")

plt.show()
```

(2) [2점] `hist()` 함수를 사용하여 2010년 인구수 분포를 보여주는 히스토그램을 출력하는 코드를 작성하시오. 그래프에는 제목, x축 및 y축의 값을 설명하는 레이블이 포함되어야 하며 나머지 부분은 기본값을 사용하면 됩니다.

```python
# 답안을 작성하세요.
plt.figure(figsize=[12, 8])
plt.hist(pop["pop10"], bins=10)
plt.title("Population Distribution in 2010")
plt.xlabel("Population")
plt.ylabel("Number of Cities")

plt.show()
```

(3) [3점] 2010년을 기준으로 경기도에서 인구가 가장 많은 다섯 개 도시를 선택하고, 선택된 도시의 인구를 막대그래프로 표현하기 위한 코드를 작성하시오. 그래프에는 제목, x축 및 y축의 값을 설명하는 레이블이 포함되어야 하며 나머지 부분은 기본값을 사용하면 됩니다.

```python
# 답안을 작성하세요.

gyeonggi_pop = pop[pop["provinces"] == "Gyeonggi"]
top5_cities_gyonggi = gyeonggi_pop.nlargest(5, "pop10")
top5_cities_gyonggi
```

```python
plt.figure(figsize=[12, 8])
plt.bar(top5_cities_gyonggi["municipality"], top5_cities_gyonggi["pop10"]/10000)
plt.title("Top 5 Most Populated Cities in Gyeonggi (2010)")
plt.xlabel("City")
plt.ylabel("Population")
plt.show()
```

(4) [3점] 2010년을 기준으로 인구가 가장 많은 다섯 개 도시를 선택하고, 해당 도시의 2000년, 2005년, 2010년 인구를 각각 다른 색상으로 표시하는 막대그래프를 그림 3과 같이 나타내는 코드를 작성하시오. 그래프의 하단에는 각 도시의 이름이 표기되어야 하며, 제목, x축 및 y축 레이블, 범례도 포함되어야 합니다. 

```python
import numpy as np
```

```python
# 답안을 작성하세요.
top5_cities = pop.nlargest(5,"pop10")
top5_cities
pop_2000 = pop['pop00']
pop_2005 = pop['pop05']
pop_2010 = pop['pop10']
x = np.arange(len(top5_cities))
```

```python
plt.figure(figsize=(12, 8))
plt.bar(x,top5_cities["pop00"]/10000, width=0.2, label = "2000years")
plt.bar(x + 0.2, top5_cities["pop05"]/10000, width=0.2, label = "2005years")
plt.bar(x + 0.4, top5_cities["pop10"]/10000, width=0.2, label = "2010years")
plt.xticks(x + 0.2, top5_cities["municipality"])
plt.legend()

plt.title("Population change in selected cities")
plt.xlabel("City")
plt.ylabel("Population (in thousabds)")
```

(5) [2점] 2000년부터 2010년까지의 기간 동안 가장 높은 인구 증가율(%)을 가진 일곱 개의 도시를 선택하기 위한 코드를 작성하시오.

```python
# 답안을 작성하세요.
pop["growth_rate"] = ((pop["pop10"] - pop["pop00"]) / pop["pop00"]) * 100
pop.head(10)
```

```python
top7_growth_cities = pop.nlargest(7, "growth_rate")
top7_growth_cities
```

(6) [2점] (5)에서 찾은 일곱 개 도시의 인구 증가율을 막대그래프 나타내는 코드를 작성하시오. 각 도시의 이름은 막대그래프 하단에 표기되어야 하며, 제목, x축 및 y축 레이블도 포함되어야 합니다.

```python
# 답안을 작성하세요.
plt.figure(figsize=(12, 8))
plt.bar(top7_growth_cities["municipality"],top7_growth_cities["growth_rate"])
plt.title("Top 7 Most Growth Cities")
plt.xlabel("City")
plt.ylabel("Growth Rate")
```
